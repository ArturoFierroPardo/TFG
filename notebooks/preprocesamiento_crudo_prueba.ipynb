```python
ruta_txt = 'C:/Users/artur/Desktop/TFG/textos/textoplano.txt'
with open(ruta_txt, 'r', encoding='utf-8') as f:
    raw = f.read()
print(raw[:500])
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(raw)
sentences = list(doc.sents)
for sent in sentences[:5]:
    print(sent)
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
print(tokens[:10])
print('\n')

for token in doc[:10]:
    print(token.text, '=>', token.pos_, '=>', token.tag_)
```

```python
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words and token.isalpha()]
print(filtered_tokens[:10])
print('\n')

doc_filtrado = nlp(" ".join(filtered_tokens))

for token in doc_filtrado[:10]:
    print(token.text, '=>', token.pos_, '=>', token.tag_)
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
stems = [stemmer.stem(token) for token in filtered_tokens]

print(stems[:20])
```

```python
lemmas = [token.lemma_ for token in doc_filtrado if token.is_alpha]
print(lemmas[:20])
```
