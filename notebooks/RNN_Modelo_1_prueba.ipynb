```python
ruta_txt = 'C:/Users/artur/Desktop/TFG/textos/textoplano.txt'
with open(ruta_txt, 'r', encoding='utf-8') as f:
    raw = f.read()
print(raw[:500])
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp(raw)
sentences = list(doc.sents)
for sent in sentences[:5]:
    print(sent)
```

```python
import nltk
from nltk.tokenize import word_tokenize
tokens = word_tokenize(raw)
print(tokens[:10])
print('\n')

for token in doc[:10]:
    print(token.text, '=>', token.pos_, '=>', token.tag_)
```

```python
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words and token.isalpha()]
print(filtered_tokens[:10])
print('\n')

doc_filtrado = nlp(" ".join(filtered_tokens))

for token in doc_filtrado[:10]:
    print(token.text, '=>', token.pos_, '=>', token.tag_)
```

```python
from nltk.stem import SnowballStemmer
stemmer = SnowballStemmer('english')
stems = [stemmer.stem(token) for token in filtered_tokens]

print(stems[:20])
```

```python
lemmas = [token.lemma_ for token in doc_filtrado if token.is_alpha]
print(lemmas[:20])
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
limited_lemmas = lemmas[:20]

tokenizer = Tokenizer()
tokenizer.fit_on_texts(limited_lemmas)
tokenizer.word_index
```

```python
encoded = tokenizer.texts_to_sequences(limited_lemmas)
encoded = [item[0] for item in encoded if len(item) > 0]
print(encoded)
```

```python
import numpy as np
from keras.utils import to_categorical
vocab_size = len(tokenizer.word_index) + 1    
x = np.array(encoded[:-1]).reshape(-1, 1)  # input: cada palabra como secuencia de un paso
y = to_categorical(encoded[1:], num_classes=len(tokenizer.word_index) + 1)
print(x, y)
```

```python
from keras.models import Sequential
from keras.layers import Embedding, LSTM, Dense

modelo = Sequential()                                           
modelo.add(Embedding(vocab_size, 10)) 
modelo.add(LSTM(50))                                        
modelo.add(Dense(vocab_size, activation='softmax'))           

modelo.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])       
```

```python
modelo.fit(x, y, epochs=120, verbose=2)
```

```python
modelo.summary()
```

```python
in_text = 'black'
print("Input:", in_text)
encoded = tokenizer.texts_to_sequences([in_text])[0]
encoded = np.array(encoded).reshape(1,1)
yhat = modelo.predict(encoded, verbose=0)
yhat_index = np.argmax(yhat)
for word, index in tokenizer.word_index.items():
	if index == yhat_index:
		print("Output:", word)
```

```python
def generate_seq(modelo, tokenizer, seed_text, n_words):
    in_text = seed_text
    result = seed_text
    
    for _ in range(n_words):
        # codificamos palabra a entero
        encoded = tokenizer.texts_to_sequences([in_text])[0]
        encoded = np.array(encoded).reshape(1, 1)  # (batch=1, timestep=1)

        # predecimos la siguiente palabra
        yhat = modelo.predict(encoded, verbose=0)
        yhat_index = np.argmax(yhat)

        # convertimos índice a palabra
        out_word = ''
        for word, index in tokenizer.word_index.items():
            if index == yhat_index:
                out_word = word
                break

        # añadimos palabra generada al texto
        in_text = out_word
        result += ' ' + out_word

    return result
```

```python
generate_seq(modelo, tokenizer, seed_text='black', n_words=10)
```
